```python
# importing the dependencies

import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

```python
# collecting the data

product_data=pd.read_csv('/content/synthetic_online_retail_data.csv')
product_data.head()
```

```python
# getting the size of the data

product_data.shape
```

```python
# selecting the important features from the data

selected_features=['category_id','category_name','product_name']
print(selected_features)
```

```python
# combining all the features

combined_features=product_data['category_id'].astype(str)+' '+product_data['category_name']+' '+product_data['product_name']
print(combined_features)
```

```python
# checking is there any null values

combined_features.isnull().sum()
```

```python
# there is no null values so we can move on
```

```python
# conveting the text data into features vector

vectorizer=TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
# finding the similaraty using cosine similaraty

similaraty=cosine_similarity(feature_vectors)
print(similaraty)
```

```python
# getting the shape of the similaraty

similaraty.shape
```

```python
# getting input from the consumer

product_name=input('Enter the product name: ')
```

```python
# creating a list of all product from the dataset

list_of_all_products=product_data['product_name'].tolist()
print(list_of_all_products)
```

```python
# finding the close match of the product

close_match=difflib.get_close_matches(product_name,list_of_all_products)
print(close_match)
```

```python
# finding the index of the product with the product name

index_of_product=product_data[product_data.product_name==close_match[0]].index[0]
print(index_of_product)
```

```python
# getting the list of similar products

similarity_score=list(enumerate(similaraty[index_of_product]))
print(similarity_score)
```

```python
# sorting the products based on their similarity score
sorted_similar_products = sorted(similarity_score, key=lambda x:x[1], reverse=True)

print('Products suggested for you :\n')
i = 1
for product in sorted_similar_products:
    index = product[0]
    product_from_dataframe = product_data[product_data.index==index]['product_name'].values[0]
    if (i<6):
        print(i, '.',product_from_dataframe)
        i+=1
```
